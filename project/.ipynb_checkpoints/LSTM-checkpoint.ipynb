## Import libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

## Parameters

```python
n_step = 3
n_feature = 6
train_test_rate = 0.2
```

## Load Dataset

```python
dataset = pd.read_csv('processed_data\p_data.csv', index_col='date', parse_dates=True)
```

```python
dataset.head()
```

## Generate Dataset

```python
from scipy.ndimage import gaussian_filter1d
from scipy.signal import medfilt
```

```python
medfilt(dataset['Ttb'], 3)[:10]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(50,10))
plt.plot((medfilt(dataset['Utb'], 3), 1.2)[:360])
plt.plot(dataset['Utb'][:360].values, c='red')
```

```python
dataset['Ttb'] = medfilt(dataset['Ttb'], 3)
dataset['Tx'] = medfilt(dataset['Tx'], 3)
dataset['Tn'] = medfilt(dataset['Tn'], 3)
dataset['Utb'] = medfilt(dataset['Utb'], 3)
dataset['Un'] = medfilt(dataset['Un'], 3)
```

```python
dataset['Ttb'][:10].values
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
dataset_scaled = scaler.fit_transform(dataset)
```

```python
dataset_scaled[:5]
```

### Splitting

* `Train-test split`

```python
def train_test_split(data, rate=0.3):
    idx = int((1 - rate)*len(data))
    return data[:idx], data[idx:]
```

```python
train_set, test_set = train_test_split(data=dataset_scaled, rate=train_test_rate)
```

* `Timestep split`

```python
def split_data(data, n_timestep):
    X, y = [], []

    length = len(data)
    for i in range(length):
        end_ix = i + n_timestep
        if end_ix > length - 1:
            break
        seq_X, seq_y = data[i:end_ix], data[end_ix]
        X.append(seq_X)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
X_train, y_train = split_data(data=train_set, n_timestep=n_step)
X_test, y_test = split_data(data=test_set, n_timestep=n_step)
```

```python
X_train[0], y_train[0]
```

```python
dataset_scaled[:4]
```

## Build model LSTM

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = keras.models.Sequential()
model.add(LSTM(units=128, activation='relu', return_sequences=True, input_shape=(n_step,n_feature)))
model.add(LSTM(units=128, activation='relu'))
model.add(Dense(units=n_feature, activation='sigmoid'))

model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
model.fit(x=X_train, y=y_train, epochs=30, verbose=0, batch_size=4, validation_data=(X_test, y_test))
```

```python
plt.figure(figsize=(19,6))
plt.plot(model.history.history['loss'])
```

```python
y_predict = model.predict(X_test)
y_predict = y_predict


y_predict = scaler.inverse_transform(y_predict)
y_real = scaler.inverse_transform(y_test)


Ttb_pred, Tx_pred, Tn_pred, R_pred, Utb_pred, Un_pred = y_predict[:,0], y_predict[:,1], y_predict[:,2], y_predict[:,3], y_predict[:,4], y_predict[:,5]

Ttb_real, Tx_real, Tn_real, R_real, Utb_real, Un_real = y_real[:,0], y_real[:,1], y_real[:,2], y_real[:,3], y_real[:,4], y_real[:,5]

print(np.shape(Ttb_pred), np.shape(Tn_real))
```
