```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

# Load Dataset

```python
dataset = pd.read_csv('data\weather_dataset.csv', index_col='date', parse_dates=True)
```

```python
dataset.head()
```

```python
dataset.tail()
```

# Data Preprocessing

```python
dataset.info()
```

### Chuyển cột mưa (R) về 0 với 1

```python
dataset = dataset.fillna(-1)
```

```python
#labling rain column
for i in range(len(dataset)):
    if  dataset['R'][i] == -1:
        dataset['R'][i] = 0
    else:
        dataset['R'][i] = 1
dataset.head()

```

```python
dataset.describe()
```

```python
#dataset.drop('R', axis=1, inplace=True)
#dataset.head()
```

### Scaling dataset

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
dataset_scaled = scaler.fit_transform(dataset)
```

```python
len(dataset_scaled)
```

# Splitting Dataset

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
rate = 0.2
X, y = dataset_scaled[:-1], dataset_scaled[1:]
pos_split = int(len(X)*(1 - rate))
X_train, X_test, y_train, y_test = X[:pos_split], X[pos_split:], y[:pos_split], y[pos_split:]
X_train[0].shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_features = 6
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(6,)))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=6, activation='sigmoid'))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
model.fit(x=X_train, y=y_train, epochs=10, verbose=0)
```

```python
#plt.plot(model.history.history['loss'])
```

```python
model.history.history['loss']
```

# Test

```python
y_real = scaler.inverse_transform(y_test)
Ttb_real, Tx_real, Tn_real, R_real, Utb_real, Un_real = y_real[:,0], y_real[:,1], y_real[:,2], y_real[:,3], y_real[:,4], y_real[:,5]

def get_MAE(model):
    y_predict = model.predict(X_test)
    y_predict = scaler.inverse_transform(y_predict)
    Ttb_pred, Tx_pred, Tn_pred, R_pred, Utb_pred, Un_pred = y_predict[:,0], y_predict[:,1], y_predict[:,2], y_predict[:,3], y_predict[:,4], y_predict[:,5]
    Ttb_MAE = np.array(tf.keras.metrics.mean_squared_error(Ttb_real, Ttb_pred))
    
    
```

```python
Ttb_MAE, Tx_MAE, Tn_MAE, R_accuracy, Utb_MAE, Un_MAE = [], [], [], [], [], []
for i in range(100):
    model.fit(x=X_train, y=y_train, epochs = 1, verbose=0)
    
    y_predict = model.predict(X_test)
    y_predict = scaler.inverse_transform(y_predict)
    Ttb_pred, Tx_pred, Tn_pred, R_pred, Utb_pred, Un_pred = y_predict[:,0], y_predict[:,1], y_predict[:,2], y_predict[:,3], y_predict[:,4], y_predict[:,5]
    
    Ttb_MAE_i = np.array(tf.keras.metrics.mean_squared_error(Ttb_real, Ttb_pred))
    Ttb_MAE.append(Ttb_MAE_i)
    
    Tx_MAE_i = np.array(tf.keras.metrics.mean_squared_error(Tx_real, Tx_pred))
    Tx_MAE.append(Tx_MAE_i)
    
    Tn_MAE_i = np.array(tf.keras.metrics.mean_squared_error(Tn_real, Tn_pred))
    Tn_MAE.append(Tn_MAE_i)
    
    BA = tf.keras.metrics.BinaryAccuracy(threshold=0.5)
    BA.update_state(R_real, R_pred)
    R_accuracy_i = BA.result().numpy()
    R_accuracy.append(R_accuracy_i*100)
    
    Utb_MAE_i = np.array(tf.keras.metrics.mean_squared_error(Utb_real, Utb_pred))
    Utb_MAE.append(Utb_MAE_i)
    
    Un_MAE_i = np.array(tf.keras.metrics.mean_squared_error(Un_real, Un_pred))
    Un_MAE.append(Un_MAE_i)
    
    
```
