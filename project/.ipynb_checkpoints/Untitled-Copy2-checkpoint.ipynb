### Outliers

```python
plt.figure(figsize=(19,6))
plt.plot(dataset.index, dataset['Ttb'], label='real')
```

```python
dataset.describe()
```

```python
dataset['Ttb'].std()
```

```python
def replace_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_threshold = Q1 - 2*column.std()
    upper_threshold = Q3 + 2*column.std()
    for i in range(len(dataset)):
        if column[i] < lower_threshold:
            column[i] = lower_threshold
        if column[i] > upper_threshold:
            column[i] = upper_threshold
```

```python
replace_outliers(dataset['Ttb'])
replace_outliers(dataset['Tx'])
replace_outliers(dataset['Tn'])
replace_outliers(dataset['R'])
replace_outliers(dataset['Utb'])
replace_outliers(dataset['Un'])
```

### Scaling dataset

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
dataset_scaled = scaler.fit_transform(dataset)
```

```python
len(dataset_scaled)
```

# Splitting Dataset

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
rate = 0.2
X, y = dataset_scaled[:-1], dataset_scaled[1:]
pos_split = int(len(X)*(1 - rate))
X_train, X_test, y_train, y_test = X[:pos_split], X[pos_split:], y[:pos_split], y[pos_split:] 
X_train[0].shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_features = 6
model = Sequential()
model.add(Dense(units=64, activation='relu', input_shape=(6,)))
model.add(Dense(units=6, activation='sigmoid'))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
model.fit(x=X_train, y=y_train, epochs = 100, verbose =1)
```

```python
plt.plot(model.history.history['loss'])
```

# Test

```python

print(X_train[0])

y_predict = model.predict(X_test)
y_predict = y_predict.T
Ttb, Tx, Tn, R, Utb, Un = y_predict[0,:], y_predict[1,:], y_predict[2,:], y_predict[3,:], y_predict[4,:], y_predict[5,:]



```
