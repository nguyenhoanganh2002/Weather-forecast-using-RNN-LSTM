`Góc nhìn rõ hơn về lượng mưa`

```python
plt.figure(figsize=(19,6))
plt.plot(dataset.index, dataset['R'])
plt.title('Lượng mưa trong ngày');
```

```python
dataset['R'].value_counts()
```

```python
rain = 0
not_rain = 0
for i in range(len(dataset)):
    if  dataset['R'][i] == -1:
        rain += 1
    else:
        not_rain += 1
print(f'Số ngày có mưa: {rain}\nSố ngày không có mưa: {not_rain}')
```

`Nhìn vào thống kê trên:`
* `Hơn 50% số ngày trong tập dữ liệu là không có mưa`
* `Luợng mưa thay đổi thất thường (ngày hôm qua không có mưa nhưng ngày hôm này có thể mưa rất nhiều)`
* `Theo tri thức từ thực tế thì để dự đoán chính xác lượng mưa thì sẽ cần có nhiều dữ kiện hơn (như mây, gió, ..)`

=> `Thay đổi giá trị của cột mưa thành 0(không có mưa) và 1(có mưa) nhằm đưa ra dự đoán chính xác hơn và tránh gây nhiễu đến         việc dự đoán các cột khác`

```python
for i in range(len(dataset)):
    if  dataset['R'][i] == -1:
        dataset['R'][i] = 0
    else:
        dataset['R'][i] = 1
```

## Seasonal decompose

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

* `Phân tích dữ liệu cột Ttb trong hai năm`

```python
fig = seasonal_decompose(dataset['Ttb'][:730]).plot()
fig.set_size_inches((20,10))
fig.savefig('image/Ttb_seasonal_decompose_two_years.png')
```

`Sau khi loại bỏ yếu tố mùa vụ và xu hướng thì còn lại phần nhiễu 'Resid'. Nhiễu nằm trong khoảng [-5,5]`

### Unnoise
 `Tiến hành khử nhiễu để thu hẹp khoảng Resid`

```python
from scipy.signal import medfilt
```

```python
unnoise_dataset = dataset.copy(deep=True)
unnoise_dataset['Ttb'] = medfilt(dataset['Ttb'], 3)
unnoise_dataset['Tx'] = medfilt(dataset['Tx'], 3)
unnoise_dataset['Tn'] = medfilt(dataset['Tn'], 3)
unnoise_dataset['Utb'] = medfilt(dataset['Utb'], 3)
unnoise_dataset['Un'] = medfilt(dataset['Un'], 3)
```

```python
plt.figure(figsize=(19, 6))
plt.plot(dataset['Ttb'][:360], c='red', label='real')
plt.plot(unnoise_dataset['Ttb'][:360], c='blue', label='unnoise')
plt.title('Ttb real vs unnoise')
plt.savefig('image/Ttb_real_vs_unnoise_one_year.png')
plt.legend();
```

```python
fig = seasonal_decompose(unnoise_dataset['Ttb'][:730]).plot()
fig.set_size_inches((20,10))
fig.savefig('image/Ttb_seasonal_decompose_two_years_after_unnoise.png')
```

`Sau khi khử nhiễu bằng bộ lọc trung vị median filter thì Resid thu về khoảng [-2.5, 2.5]`
